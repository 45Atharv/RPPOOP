```python
class Bank():
  def __init__(self,name,acc_number,acc_type,balance=100):
    self.name = name
    self.acc_number = acc_number
    self.acc_type = acc_type
    self.balance = balance

  def deposit(self,amount):
    self.balance = self.balance + amount 
    print("Amount deposited successfully")


  def withdraw(self,amount):
    if (self.balance < amount):
      print("Insufficient Balance")
    else:
      self.balance= self.balance-amount
      print("Account debited successfully")

  def display(self):
    print("Account Holder Name: ",self.name)
    print("\tAccount Number: ",self.acc_number)
    print(" \tAccount Type: ",self.acc_type)
    print("\tCurrent balance:", self.balance)


print("Welcome To COEP bank\n")
name = input("Enter your registered name : ")
number = int (input("Enter your account number :"))
typ = input("Enter your type of account : ")



  
print("Enter 1 for Deposition of amount\n")
print("Enter 2 for Withdrawal of amount\n")
print("Enter 3 for to see account status\n")

x =int (input("Enter your choice : "))

if (x==1):
  amt = int(input("Enter amount to be deposited :"))

  acc1 = Bank(name , number ,typ , balance=100)
  acc1.display()
  acc1 .deposit(amt)
  acc1.display()

elif ( x==2):
  amt = int(input("Enter amount to be withdrawn :"))

  acc1 = Bank(name , number ,typ , balance=100)
  acc1.display()
  acc1 .withdraw(amt)
  acc1.display()

elif (x==3):
  acc1 = Bank(name , number ,typ , balance=100)
  acc1.display()
  
```
